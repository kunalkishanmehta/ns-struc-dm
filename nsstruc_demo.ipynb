```python
import numpy as np # import packages
import matplotlib.pyplot as plt
```

```python
# download and install the code

%%bash
# make this cell an instance of a terminal

git clone -b enthalpy https://github.com/landryp/ns-struc.git # download code repository from https://github.com/landryp/ns-struc/tree/enthalpy

mkdir /opt/bin /opt/lib/ /opt/lib/python2.7 /opt/lib/python2.7/site-packages # create install directory and make it findable
export PATH=$PATH:/opt/bin
export PYTHONPATH=$PYTHONPATH:/opt/lib/python2.7/site-packages

cd /content/ns-struc # install code
python2 setup.py install --prefix /opt/

echo -e '\n *** \n'

getnsprops -h # this is the executable from ns-struc that solves the TOV equations to compute neutron star observables; print its help string to see the syntax for calling it

echo -e '\n *** \n'
```

```python
# calculate neutron star observables for a given equation of state

%%bash
# this cell is a separate terminal instance

export PATH=$PATH:/opt/bin # the code doesn't need to be reinstalled in each terminal, but we do need to point to the install directory like this each time
export PYTHONPATH=$PYTHONPATH:/opt/lib/python2.7/site-packages

getnsprops mpa1.csv -p M,R,Lambda -m 1.,3. -d /content/ns-struc/ -o /content/ -v # call the executable that computes the neutron star observables
# getnsprops [eos table] -p [observables to calculate] -m [mass range to calculate them for] 
#-d [directory where the eos table lives] -o [directory where the table of observables should be output]
```

```python
mrl_data_hqc18 = np.genfromtxt('/content/macro-hqc18.csv',delimiter=',',names=True,dtype=None) # load the table of neutron star observables that was created above
mrl_data_sly230a = np.genfromtxt('/content/macro-sly230a.csv',delimiter=',',names=True,dtype=None) # load the table of neutron star observables that was created above
mrl_data_sly = np.genfromtxt('/content/macro-sly.csv',delimiter=',',names=True,dtype=None) # load the table of neutron star observables that was created above
mrl_data_mpa1 = np.genfromtxt('/content/macro-mpa1.csv',delimiter=',',names=True,dtype=None) # load the table of neutron star observables that was created above

plt.figure(figsize= (11,7))
plt.plot(mrl_data_mpa1['R'],mrl_data_mpa1['M'], label = 'mpa1') # plot the mass-radius relation
plt.plot(mrl_data_hqc18['R'],mrl_data_hqc18['M'],label ='hqc18') # plot the mass-radius relation
plt.plot(mrl_data_sly230a['R'],mrl_data_sly230a['M'], label  ='sly230a', c = 'k') # plot the mass-radius relation
plt.plot(mrl_data_sly['R'],mrl_data_sly['M'], label = 'sly') # plot the mass-radius relation

plt.xlabel(r'$R$ [km]', size = 16) # add plot labels, etc.
plt.ylabel(r'$M$ [$M_\odot$]', size = 16)
plt.xlim(8.,16.)
plt.ylim(1.,3.)
plt.legend(fontsize = 16, title='Equation of State',title_fontsize=18 )
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('Mass-radius.png')

plt.show()
```

```python
plt.figure(figsize= (11,7))
plt.plot(mrl_data_mpa1['Lambda'],mrl_data_mpa1['M'], label = 'mpa1') # plot the mass-radius relation
plt.plot(mrl_data_hqc18['Lambda'],mrl_data_hqc18['M'],label ='hqc18') # plot the mass-radius relation
plt.plot(mrl_data_sly230a['Lambda'],mrl_data_sly230a['M'], label  ='sly230a', c = 'k') # plot the mass-radius relation
plt.plot(mrl_data_sly['Lambda'],mrl_data_sly['M'], label = 'sly') # plot the mass-radius relation

plt.xlabel(r'$\Lambda$', size = 16) # add plot labels, etc.
plt.ylabel(r'$M$ [$M_\odot$]', size = 16)
# plt.xlim(8.,16.)
# plt.ylim(1.,3.)
plt.legend(fontsize = 16, title='Equation of State',title_fontsize=18 )
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('Mass-tidaldeform.png')

plt.show()
```

```python
plt.figure(figsize= (11,7))
plt.plot(mrl_data_mpa1['rhoc'],mrl_data_mpa1['M'], label = 'mpa1') # plot the mass-radius relation
plt.plot(mrl_data_hqc18['rhoc'],mrl_data_hqc18['M'],label ='hqc18') # plot the mass-radius relation
plt.plot(mrl_data_sly230a['rhoc'],mrl_data_sly230a['M'], label  ='sly230a', c = 'k') # plot the mass-radius relation
plt.plot(mrl_data_sly['rhoc'],mrl_data_sly['M'], label = 'sly') # plot the mass-radius relation

plt.xlabel(r'$\rho_c$', size = 16) # add plot labels, etc.
plt.ylabel(r'$M$ [$M_\odot$]', size = 16)
# plt.xlim(8.,16.)
# plt.ylim(1.,3.)
plt.legend(fontsize = 16, title='Equation of State',title_fontsize=18 )
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xscale('log')
plt.savefig('Mass-rhoc.png')

plt.show()
```
